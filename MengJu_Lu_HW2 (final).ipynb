Missing Data Processing

```python
#G	AB	R	H	2B	3B	HR	RBI	SB	CS	BB	SO	IBB	HBP	SH	SF	GIDP

if df["G"].isna().sum()>0:
    df["G"]=df["G"].fillna(value=0)
if df["AB"].isna().sum()>0:
    df["AB"]=df["AB"].fillna(value=0)
if df["R"].isna().sum()>0:
    df["R"]=df["R"].fillna(value=0)
if df["H"].isna().sum()>0:
    df["H"]=df["H"].fillna(value=0)
if df["2B"].isna().sum()>0:
    df["2B"]=df["2B"].fillna(value=0)
if df["3B"].isna().sum()>0:
    df["3B"]=df["3B"].fillna(value=0)
if df["HR"].isna().sum()>0:
    df["HR"]=df["HR"].fillna(value=0)
if df["RBI"].isna().sum()>0:
    df["RBI"]=df["RBI"].fillna(value=0)
if df["SB"].isna().sum()>0:
    df["SB"]=df["SB"].fillna(value=0)
if df["CS"].isna().sum()>0:
    df["CS"]=df["CS"].fillna(value=0)
if df["BB"].isna().sum()>0:
    df["BB"]=df["BB"].fillna(value=0)
if df["SO"].isna().sum()>0:
    df["SO"]=df["SO"].fillna(value=0)
if df["IBB"].isna().sum()>0:
    df["IBB"]=df["IBB"].fillna(value=0)
if df["HBP"].isna().sum()>0:
    df["HBP"]=df["HBP"].fillna(value=0)
if df["SH"].isna().sum()>0:
    df["SH"]=df["SH"].fillna(value=0)
if df["SF"].isna().sum()>0:
    df["SF"]=df["SF"].fillna(value=0)
if df["GIDP"].isna().sum()>0:
    df["GIDP"]=df["GIDP"].fillna(value=0)
df
```
